```python
df = pd.read_csv("../dataset/adult/adult.csv")
```

```python
df.head()
```

```python
df1 = df[["sex", "age"]]
df1.head()
```

```python
df['sex'].unique()
```

```python
df_list = []
for cate in df1["sex"].unique():
    df2 = df1[df1["sex"]==cate]["age"]
    df_list.append(df2)
```

```python
df_list[0]
```

```python
df_list[1]
```

```python
import plotly.graph_objects as go

import numpy as np
names = df1["sex"].unique()
fig = go.Figure()
```

```python
for hist_data, name in zip(df_list, names):
    fig.add_trace(go.Histogram(x=hist_data, name=name) )
    
```
